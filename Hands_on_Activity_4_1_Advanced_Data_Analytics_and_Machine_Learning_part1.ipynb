## **Simple Linear Regression**

```python
Training_ID = training[['PassengerId']]  #PassengerId = Passengers(featured)
Training_Survival = training['Survived']  #1.0 Alive || 0.0 dead(targets)
```

```python
model = LinearRegression()
model.fit(Training_ID, Training_Survival)
```

```python
y_pred = model.predict(Training_ID)
```

```python
plt.scatter(Training_ID, Training_Survival)
plt.plot(Training_ID, y_pred, color='red')
plt.xlabel('Passengers')
plt.ylabel('Survived')
plt.title('Simple Linear Regression')
plt.show()
```
